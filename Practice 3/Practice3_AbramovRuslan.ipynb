### Задание 22. Визуализировать данные вашего скаченного DataSet в стиле Столбчатые диаграммы (подписать оси, заголовок)

```python
df_filtered = df2.head(100) # Пример на первых 100 человек, чтобы увидеть разницу

df2_avg = df_filtered.groupby('Age')['Weight'].mean().reset_index() # У одного возраста может быть несколько значений веса, потому берём среднее
plt.figure(figsize=(10, 6))
plt.bar(df2_avg['Age'], df2_avg['Weight'], color='blue')
plt.title('Средний вес людей с сердечными заболеваниями в зависимости от возраста (100 человек)', fontweight='bold', fontsize=10)
plt.xlabel('Возраст', fontweight='bold', fontsize=15)
plt.ylabel('Средний вес (в килограммах)', fontweight='bold', fontsize=15)
plt.show()
```

```python
df2_avg = df2.groupby('Age')['Weight'].mean().reset_index() #Среднее значение веса на каждый возраст
plt.figure(figsize=(10, 6))
plt.bar(df2_avg['Age'], df2_avg['Weight'], color='blue')
plt.title('Средний вес людей с сердечными заболеваниями в зависимости от возраста (Все данные)', fontweight='bold', fontsize=10)
plt.xlabel('Возраст', fontweight='bold', fontsize=15)
plt.ylabel('Средний вес (в килограммах)', fontweight='bold', fontsize=15)
plt.show()
```
